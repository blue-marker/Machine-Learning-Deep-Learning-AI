```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("salary.csv",index_col=0)
df
```

```python
df.info()
```

```python
df.nunique()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for col in df:
    if col in ('Workclass','education','marital-status','occupation','relationship','race','sex','native-country'):
        print(col)
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col])
    else:
        continue
```

```python
df["fnlwgt"] = round(np.cbrt(df["fnlwgt"]))
df["capital-gain"] = round(np.cbrt(df["capital-gain"]))
df["capital-loss"] = round(np.cbrt(df["capital-loss"]))
df
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,
                                                 random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression()
```

```python
df["Income"].value_counts()
# the data set is imbalance with negative class(salary< 50k) having more in
# proportion compared to positive class(salary >50k)
```

```python
log.fit(X_train,y_train)
```

```python
log.score(X_test,y_test)
```

```python
log.coef_
```

```python
y_pred = log.predict(X_test)
y_pred
```

### Evaluation Metrics

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
tn, fp, fn, tp = confusion_matrix(y_test,y_pred).ravel()
```

```python
print(tp, fn)
print(fp, tn)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
precision_score(y_test,y_pred)
```

```python
recall_score(y_test,y_pred)
```

```python
f1_score(y_test,y_pred)
```

```python
from sklearn.metrics import roc_curve

fpr, tpr, threshold = roc_curve(y_test, y_pred)
```

```python
from sklearn.metrics import roc_auc_score

auc = roc_auc_score(y_test, y_pred)
print(auc)
```
